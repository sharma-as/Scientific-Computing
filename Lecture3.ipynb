## Functions

+ Write a fucntion to compute the first and second order derivatives of a function using the central difference approximation.

```python
def derivatives(f,x,dx):
    df = (f(x+dx)-f(x-dx))/(2*dx)
    ddf = (f(x+dx)-2*f(x)+f(x-dx))/dx**2
    return df, ddf
```

Let us now use the fucntion to determine the derivatives sin(x) at x = 0.5

```python
from math import sin
df, ddf = derivatives(sin,0.5,1e-4)
print(f"First derivative : {df}, Second derivative : {ddf}")
```

You can also assign default values to function argumnets.

```python
def derivatives_V2(f,x,dx=1e-5):  # assign default value for dx
    df = (f(x+dx)-f(x-dx))/(2*dx)
    ddf = (f(x+dx)-2*f(x)+f(x-dx))/dx**2
    return df, ddf
```

```python
from math import atan
df, ddf = derivatives_V2(atan,0.5)  # uses default of dx = 1e-5
print(df, ddf)
```

```python
derivatives_V2(atan,0.5,0.01) # User-specified value of dx
```

# Operations on Arrays

Arithmetic operators work differently on arrays than they do on tuples and lists - the operation is *broadcast* to all the elements of the array; that is; the operation is applied to each element in the array.

```python
a = np.array([0.0, 1.0, 4.0, 9.0, 16.0, 25.0])
```

```python
a / 25
```

```python
a - 10
```

```python
np.sqrt(a)
```

# Vectorizing algorithms

Sometimes the broadcasting properties of the mathematical functions in the numpy module can be used to replace loops in the code. This procedure is known as **vectorization**. 

Consider for example, the expression <br>
$$ s = \sum_{i=0}^{100} \sqrt{\frac{i\pi}{100}} \sin\bigg(\frac{i\pi}{100}\bigg) $$

The direct approach is to evaluate the sum in a loop, resulting in the following scalar code

```python
import math
```

```python
%%timeit
x = 0.0
my_sum = 0.0
for i in range(101):
    my_sum = my_sum + math.sqrt(x)*math.sin(x)
    x = x + 0.01*math.pi
```

The vectorized version of the algorithm using the numpy fucntions is,

```python
import numpy as np
x = np.arange(0.0,1.001*np.pi, 0.01*pi)
x
```

```python
np.sum(np.sqrt(x)*np.sin(x))
```

```python
%%timeit
np.sum(np.sqrt(x)*np.sin(x))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = np.linspace(0, 10, 100)
y = 4 + 1 * np.sin(2 * x)
x2 = np.linspace(0, 10, 25)
y2 = 4 + 1 * np.sin(2 * x2)

# plot
fig, ax = plt.subplots()

ax.plot(x2, y2 + 2.5, 'x', markeredgewidth=2)
ax.plot(x, y, linewidth=2.0)
ax.plot(x2, y2 - 2.5, 'o-', linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
